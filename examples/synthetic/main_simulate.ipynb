```python
import os, sys

cwd = os.getcwd()
grandparent_dir = os.path.abspath(os.path.join(cwd, os.pardir, os.pardir))
if grandparent_dir not in sys.path:
    sys.path.append(grandparent_dir)

import itertools
from joblib import Parallel, delayed

from examples.synthetic.implementation.sim_est_get_results import run_sims
```

```python
# factors = [1, 3]
factors = [3]
# sample = [50, 200]
sample = [200]
missings = [0, 0.3]
n_vars = [10, 100]
# non_linear = [False, True]
non_linear = [True]
# ρ = [0.5, 0.9]
ρ = [0.5]
α = [0, 0.5]

all_combs = itertools.product(*[factors, sample, missings, n_vars, non_linear, ρ, α])
all_combs = list(all_combs)
print("n configurations: ", len(all_combs))
```

```python
# check which ones have not been run
all_files = os.listdir()
list_done = []
for this_file in all_files:
    if this_file[-4:] == ".csv":
        a_ = this_file[:-4].rsplit("_")
        for j in range(2):
            a_[j] = int(a_[j])
        if a_[2] == "0":
            a_[2] = 0
        else:
            a_[2] = float(a_[2])
        a_[3] = int(a_[3])
        a_[4] = a_[4] == "True"
        for j in range(5, 7):
            a_[j] = float(a_[j])
        list_done.append(tuple(a_))
conf_left = []
for j in all_combs:
    if j not in list_done:
        conf_left.append(j)
print("configurations left: ", len(conf_left))
```

```python
n_cpu = 8  # -1
Parallel(n_jobs=n_cpu, verbose=2)(
    delayed(run_sims)(this_comb) for this_comb in conf_left
)
```
